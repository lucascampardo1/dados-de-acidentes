```python
import pandas as pd 
df = pd.read_csv('cat_acidentes.csv', sep = ';')
```

```python
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)

df.dropna(subset=['latitude', 'longitude'])
```

```python
import folium
from folium.plugins import HeatMap

mapa = folium.Map(location=[-30.1, -51.15], zoom_start=11)

coordenadas = list(zip(df['latitude'], df['longitude']))
mapa_de_calor = HeatMap(coordenadas, radius=15, blur=12)
mapa.add_child(mapa_de_calor)

# SOMENTE MAPA DE CALOR
mapa.save('mapas/mapa_de_calor_acidentes.html')
```

```python
from folium.plugins import MarkerCluster

mapa = folium.Map(location=[-30.1, -51.15], zoom_start=11)
cluster = MarkerCluster(coordenadas)
mapa.add_child(cluster)

# MAPA COM CLUSTERS
mapa.save('mapas/mapa_cluster.html')
```
